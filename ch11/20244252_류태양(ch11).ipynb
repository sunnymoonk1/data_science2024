```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 11)
y1 = 2 + 2 * x
y2 = 1 + 2 * x + x ** 2

plt.plot(x, y1, marker = 'o', color = 'r', linestyle = 'dashed', label = 'linear')
plt.plot(x, y2,'k^:', label = 'quad')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('equation')
plt.show()
```

```python
plt.scatter(x, y1, marker = '^')
plt.scatter(x, y2, marker = '*')
plt.show()
```

## 요점정리
### 1) 시각화(visualization)은 무엇이고 왜 필요한가?
- 점이나 선, 막대 그래프 등의 시작적 이미지를 사용하여 데이터를 화면에 표시하는 것이다.
- 시각화된 정보는 비교 분석이나 관계 이해들 데이터에 내재하는 패턴을 쉽게 알아볼 수 있다.
- 교육이나 홍보 등의 분야에서도 의사 소통의 도구로도 사용한다. 

### 2) 차트의 종류에는 어떤 것들이 있나? 종류와 사용하는 함수를 찾아보자.
- 선 차트 (Line Chart) :  plt.plot()
- 산점도 (Scatter Chart) : plt.scatter()
- 막대 차트 (Bar Chart) : plt.bar(), plt.barh() 
- 파이 차트 (Pie Chart) : plt.pie()
- 히스토그램 (Histogram) : plt.hist()
- 상자 차트 (Box Chart) : plt.boxplot()

### 3) 히스토그램은 무엇인가?
- 주어진 자료를 몇 개의 구간으로 나누고 각 구간의 도수(frequency)조사하여 나타낸 막대 그래프이다.

### 4) 히스토그램을 그리는 과정을 찾아서 정리해 보자(option).
- matplotlib에서 pyplot 패키지 hist() 함수를 호출하면 된다.
- 호출하고 나서 두 개의 매개변수를 입력하고 plt.show()를 사용하여 화면을 통해 볼 수 있다.

### 5) 상자차트(box chart)차트에서 표시되는 값들은 어떠한 의미를 갖는지 찾아보자.
- 상자 차트 (box chart)는 데이터의 최대, 최소, 중간값과 사분위 수 등을 가시하는 방법이다.
- 중앙값 (median) : 중심으로 상위 25%와 하위 25%의 값이 모여 있는 구간을 표시한다.
- 위스커 (whisker) : 수염이라고도 하며, 아래쪽 위스커에서 위쪽 위스커까지의 범위가 분포하는 범위에서 최대값과 최소값을 표시한다.
- 이상치 (outlier) : 분포하는 범위 밖의 데이터를 말한다.
- 사분범위, IQR (inter-quartile range) : 상단에서 하단을 뺀 값이며, 1.5배 이상 높거나 낮은 값은 이상치로 간주한다.

```python
import matplotlib.pyplot as plt
import numpy as np

month = [7, 8, 9, 10, 11, 12]
new_user = [456, 492, 578, 599, 670, 854]

plt.figure(figsize=(6, 9))
plt.plot(month, new_user)
plt.title("Daehan company speeda net new customers")
plt.xlabel("Month")
plt.ylabel("User")
plt.show()
```

```python
fig, ax = plt.subplots(1,4,figsize=(10,3))
ax[0].bar(month, new_user)
ax[0].set_title("Bar chart")
ax[1].plot(month, new_user)
ax[1].set_title("Line chart")
ax[2].scatter(month, new_user, marker = '^')
ax[2].set_title("Scatter chart")
ax[3].barh(month, new_user)
ax[3].set_title("Horizontal Bar chart")
plt.tight_layout()
plt.show
```

```python
import numpy as np

theta = np.linspace(0, 2 * np.pi * 6, 1000)
cos = 20 * np.cos(theta)
sin = theta * np.sin(theta)

plt.figure(figsize=(10, 3))
plt.plot(theta, cos, 'b--', label ='cos')
plt.plot(theta, sin, 'r-', label ='sin')
plt.legend()
plt.savefig('sin-cos.png', dpi=150)
plt.show
```

```python
dach = {'length' : [77,78,85,83,73,77,73,80],
       'height' : [25,28,19,30,21,22,17,35]}

samo = {'length' : [75,77,86,86,79,83,83,88],
       'height' : [56,57,50,53,60,53,49,61]}

malt = {'length' : [34,38,38,41,30,37,41,35],
       'height' : [22,25,19,30,21,24,28,18]}
```

```python

plt.scatter(dach['length'], dach['height'], color = 'r', label = 'dach')
plt.scatter(samo['length'], samo['height'], color = 'b', marker ='s', label = 'samo')
plt.scatter(malt['length'], malt['height'], color = 'g', marker = '^', label = 'malt')
plt.legend()
plt.xlabel('length')
plt.ylabel('height')
plt.title('Dog size')
plt.show
```

```python
fig, ax = plt.subplots(1,3,figsize = (9,3))
ax[0].scatter(dach['length'], dach['height'], color = 'r', label = 'dach')
ax[0].set_title('dach')
ax[0].set_xlabel('length')
ax[0].set_ylabel('height')

ax[1].scatter(samo['length'], samo['height'], color = 'b', marker ='s', label = 'samo')
ax[1].set_title('samo')
ax[1].set_xlabel('length')
ax[1].set_ylabel('height')

ax[2].scatter(malt['length'], malt['height'], color = 'g', marker = '^', label = 'malt')
ax[2].set_title('malt')
ax[2].set_xlabel('length')
ax[2].set_ylabel('height')

plt.tight_layout()
plt.show()
```
