## 요점정리
### 1) 판다스가 엑셀 보다 우수한 점은 무엇인가?
- 데이터 용량이 한계 없이 대규모로 처리할 수 있다.
- 데이터 분석에서 더 많은 분석 기능을 가지고 있다.
- 다양한 부분에서 엑셀은 기본적인 기능을 보유하고 있지만, 판다스는 데이터와 관련한 기능에서 더욱 많이 가지고 있어 데이터를 관리할 때 편리하다.

### 2) 판다스에는 Series와 DataFrame 두가지 형식이 있다. 차이점을 찾아보자.
- Series(시리즈)는 1차원 구조로 레이블이 붙어있는 1차원 벡터이다.
- DataFrame(데이터프레임)은 2차원 구조로 행과 열로 되어있는 2차원 테이블로 각 열은 시리즈로 되어있다.

### 3) 판다스 데이터의 인덱싱과 슬라이싱 방법을 정리해 보자.
- 기본적으로 인덱싱은 열(컬럼) 인덱싱과 행 인덱싱이 있는데 보통 열 인덱싱을
df["column_name"]으로 정하고 행 인덱싱을 df[0:5]로 선택하여 인덱싱을 할 수 있다.

- 슬라이싱은 3가지의 방법이 있다. 앞 5개는 head(), 뒤 5개는 tail()를 사용하고 만약 행이 많을 때는 df.loc[인덱스]를 사용하여 원하는 행을 선택할 수 있다. 만약 행이 아닌 열을 선택하고 싶을 때는 ['population']을 뒤에 입력하면 된다.

### 4) 데이터 분석에서 결손 값(missing values)를 처리하는 중요하다. 어떠한 방법이 있는지 알아 보자.
- isna() : 결손값이 있는지 유무를 확인할 때 사용하는 메소드이다.
- dropna() : 결손값을 가지고 있는 행이나 열을 삭제한다.
- fllna() : 특정한 값을 다른 값으로 교체한다.

```python
car_dict = {'name'   : ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'horse power' : [130, 250, 190, 300, 210, 220, 170],
    'weight' : [1.9, 2.6, 2.2, 2.9, 2.4, 2.3, 2.2],
    'efficiency' : [16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2] }
car_P = pd.DataFrame(car_dict)
car_P = car_P.set_index('name')
car_P
```

```python
car_P[car_P['horse power'] == car_P['horse power'].max()]
```

```python
car_P['hp x mile'] = car_P['horse power'] * car_P['efficiency']
car_P.head(3)
```

### 12.3

```python
car_dict = {'name' : ['A', 'B', 'C', 'D'],
    'horse power' : [120, 220, 120, 200],
    'weight' : [1.9, 2.1, 1.5, 2.9],
    'efficiency' : [18.3, 19.2, 21.1, 17.3]}

car_Q = pd.DataFrame(car_dict).set_index('name')
car_Q
```

```python
car_P.head()
```

```python
car_PQ = pd.concat([car_P, car_Q], axis = 0)
car_PQ
```

```python
car_PQ = pd.concat([car_P, car_Q], axis = 1)
car_PQ
```

```python
car_PQ["com"] = ['P'] * len(car_P) + ['Q'] * len(car_Q)
car_PQ
```

```python
car_PQ['hp x mile'] = car_PQ['horse power'] * car_PQ['efficiency']
car_PQ.head()
```

```python
car_PQ.groupby('com')['hp x mile'].mean()
```
